```python
def sort(l):
  for i in range(len(l)):
    for j in range(len(l)-1):
      if l[j]>l[j+1]:
        temp=l[j]
        l[j]=l[j+1]
        l[j+1]=temp

l1=[5,3,7,2,8,4,9]
print(l1)
sort(l1)
print(l1)
```

```python
class Student:
  def __init__(self, id, age):
    self.id=id
    self.age=age

def sort(l, compFun): #compFunc 0- equal,
  #-1 left bigger than right, 1 = left is smaller than right
  '''
  compFun takes two argument(o1, o2) and its hould compare two values
  return >0 if o1<o2
  return 0 if o1==o2
  return <0 if o1>o2
  '''
  for i in range(len(l)):
    for j in range(len(l)-1):
      if compFun(l[j],l[j+1])<0:
        temp=l[j]
        l[j]=l[j+1]
        l[j+1]=temp

def myComFun(s1, s2):
  if s1.age==s2.age:
    return 0
  elif s1.age > s2.age:
    return -1
  else:
    return 1

def myCompFunc1(s1, s2):
  return s2.age-s1.age

f1=myCompFunc1
f2=lambda s1, s2: s2.age-s1.age

s1=Student(101, 21)
s2=Student(102, 22)
s3=Student(103, 20)
s4=Student(104, 24)
s5=Student(105, 23)

l1=[s2, s1, s5, s3, s4]
print("Before sort..")
for i in range(len(l1)):
  print(l1[i].id, l1[i].age)

print("after sort...")
sort(l1, myCompFunc1)
for i in range(len(l1)):
  print(l1[i].id, l1[i].age)

print("after sort2...")
sort(l1, lambda s1, s2: s2.age-s1.age)
for i in range(len(l1)):
  print(l1[i].id, l1[i].age)
```
