**1. Series**

```python
#series is one dimensioanl data structure
import pandas as pd
import numpy as np

sr=pd.Series([1,2,3,4,5,6])
print(type(sr))
print(sr)

#create series from list
sr=pd.Series([1,2,3,4,5,6],
             name="values", index=['001', '002', '003', '004', '005', '006'])
print(sr)

#create series from ndarray
nparr=np.array([1,2,3,4,5,6])
print(type(nparr))
print(nparr)

sr=pd.Series(nparr)
print(sr)

#create serires from dictionary
data = {'a' : 0., 'b' : 1., 'c' : 2.}
print(type(data))
print(data)
s = pd.Series(data)
print(s)

#create serries from scalar
sr=pd.Series(6, index=[0,1,2,3])
print(sr)
```

**11. Concatenation**

```python
df1=pd.DataFrame([[1,2,3],[4,5,6],[7,8,9]], columns=['c1','c2','c3'])
print(df1)
df2=pd.DataFrame([[11,12,13],[14,15,16],[17,18,19]], columns=['c1','c2','c3'])
print(df2)

print(pd.concat([df1,df2], axis=0)) #defualt rowwise concatenation
print(pd.concat([df1,df2], axis=1)) #side by side

df1=pd.DataFrame([[1,2,3],[4,5,6],[7,8,9]], columns=['c1','c2','c3'])
df2=pd.DataFrame([[11,12,13],[14,15,16],[17,18,19]], columns=['c1','c2','c3'])
df2.index=[1,2,3]

print(pd.concat([df1,df2], axis=1, join='inner')) #side by side
print(pd.concat([df1,df2], axis=1, join='outer'))

df1=pd.DataFrame([[1,2,3],[4,5,6],[7,8,9]], columns=['c1','c2','c3'])
df2=pd.DataFrame([[9,12,13],[14,15,16],[17,18,19]], columns=['c3','c4','c5'])
print(pd.concat([df1,df2], axis=0, join='inner'))
print(pd.concat([df1,df2], axis=0, join='outer'))
```

**16. Function**

```python
#how to call user defined function on dataframe
#1. pipe - Entire dataframe is passed as argument to the function
def myfunc1(x): #argument is DataFrame
  print(type(x))
  x['c1']=x['c1']+2
  return x

def myfunc2(x):
  x['c2']=x['c2']-4
  return x

def myfunc3(x):
  x['c3']=x['c3']/2
  return x

df1=pd.DataFrame([[1,2,3],[4,5,6],[1,8,9], [10,11,12]], columns=['c1','c2','c3'])
print(df1)
print(df1.pipe(myfunc1).pipe(myfunc2).pipe(myfunc3))

#2. call funciton for column or row
def mycolumfunc(col):
  print(type(col))
  total=0
  for x in col:
     total+=x
  return total

def myrowfunc(row):
  total=0
  for x in row:
     total+=x
  return total

print(df1.apply(mycolumfunc))
print(df1.apply(myrowfunc, axis=1))

#3. call funciton for each element
def myelementfunc(x):
  print(type(x))
  return x*x

print(df1.map(myelementfunc))
```

**17. Window Function**

```python
import pandas as pd

df1=pd.DataFrame([[1,2,3],[4,5,6],[1,8,9], [10,11,12]], columns=['c1','c2','c3'])
print(df1)
print(df1.rolling(window=2).sum())
print(df1.rolling(window=2, axis=1).sum())

#transpose of dataframe
print(df1.T)
print(df1.T.rolling(window=2).sum().T)

print(df1)
print(df1.rolling(window=3).sum())
#for triang window of size 3 the weight is [1,2,1]
print(df1.rolling(window=3, win_type='triang').sum())

#expanding window
print(df1)
print(df1.expanding(min_periods=2).sum())
```

**18. Aggeration**

```python
df=pd.DataFrame([[1,2,3],[4,5,6],[1,8,9], [10,11,12]], columns=['c1','c2','c3'])
print(df)
print(df.sum()) #aggregate function but we are callign one function at time

print(df.mean())
print(df.count())

#if i want to have more than one aggregation result on a dataframe
print(df.agg(['sum', 'mean','count']))
print(df.agg(['sum', 'mean','count'], axis=1))
print(df.agg({'c1':['sum'], 'c2':['mean'],'c3':['count']}))

print(df.rolling(window=3).agg(['sum', 'mean','count']))
```

**19. Merge/Join**

```python
df1=pd.DataFrame([[1,11,3],[2,12,6],[3,13,9], [4,14,12]], columns=['c1','c2','c3'])
print(df1)

df2=pd.DataFrame([[2,11,13],[3,13,16],[4,14,19], [5,15,112]], columns=['c1','c2','c6'])
print(df2)


#merge is qual to database join
#by default it uses columsn iwth same name to join the records
#by default it perfomrs inner join
print(df1.merge(df2))

print(df1.merge(df2, on='c2')) #inner join
print(df1.merge(df2, on='c2', how='left'))
print(df1.merge(df2, on='c2', how='right'))
print(df1.merge(df2, on='c2', how='outer'))

#join using index values instead of column values
print(df1.merge(df2, left_index=True, right_index=True))
```

**21. Categorical Data**

**22. Missing Data**

```python
#checking missing values
import pandas as pd
import numpy as np

df=pd.DataFrame([[1,np.nan,3],[2,12,6],[3,13,9], [4,np.nan,12]], columns=['c1','c2','c3'])
print(df)
print(df.isna()) # retunr True/False for each cell
print(df.notna())
print(df.count())
df=pd.DataFrame([[1,np.nan,3],[2,np.nan,6],[3,np.nan,9], [4,np.nan,12]], columns=['c1','c2','c3'])
print(df.count()) #pandas treats null as 0.
#If all are null then it return null. But it return 0

#filling missing value
df=pd.DataFrame([[1,np.nan,3],[2,12,6],[3,13,9], [4,np.nan,12]], columns=['c1','c2','c3'])
print(df)
print(df.fillna(0))
print(df.fillna(df.mean()))

#missing value drop
print(df)
print(df.dropna()) #drop rows that has null value
print(df.dropna(axis=1))

df=pd.DataFrame([[np.nan,np.nan,np.nan],[2,12,6],[3,13,9], [4,np.nan,12]], columns=['c1','c2','c3'])
print(df.dropna(how='all'))
print(df.dropna(thresh=3))

print(df)
print(df.dropna(subset=['c2']))

#forward fill
print(df)
print(df.ffill())

#backward fill
print(df.bfill())

#limiting ffil
df=pd.DataFrame([[1,np.nan,np.nan],[2,12,6],[3,13,9], [4,np.nan,12]], columns=['c1','c2','c3'])
print(df.ffill(limit=1, axis=1))

#limiting bfill
df=pd.DataFrame([[1,np.nan,3],[2,np.nan,6],[3,np.nan,9], [4,5,12]], columns=['c1','c2','c3'])
print(df)
print(df.bfill(limit=2))

#interppolate missing value
df=pd.DataFrame([[1,2,3],[2,np.nan,6],[3,np.nan,9], [4,5,12]], columns=['c1','c2','c3'])
print(df)
print(df.interpolate())
print(df.interpolate(axis=1))

```

**23. Duplicate Data**

```python
df=pd.DataFrame([[1,2,3],[4,5,6],[1,2,3], [4,5,12]], columns=['c1','c2','c3'])
print(df)
print(df.duplicated())

#remove duplicated records
#option 1
print(df[df.duplicated()!=True])

#option 2
print(df.drop_duplicates())

df=pd.DataFrame([[1,2,3],[4,5,6],[7,2,3], [4,5,12]], columns=['c1','c2','c3'])
print(df)
print(df.drop_duplicates(subset=['c1','c2']))
print(df.drop_duplicates(subset=['c2','c3']))

#counting on duplicate data
print(df)
print(df.nunique())
print(df.value_counts())
print(df.nunique())

#dupliacte index
df=pd.DataFrame([[1,2,3],[4,5,6],[7,2,3], [4,5,12]], columns=['c1','c2','c3'])
df.index=['001','002','001','003']
print(df)
print(df.index.is_unique)
print(df.index.duplicated())
print(df[df.index.duplicated()])
```

**24. Group By**

```python
df=pd.DataFrame([[1,2,3],[4,5,6],[7,2,3], [4,5,12]], columns=['c1','c2','c3'])
df.index=['001','002','001','003']
print(df)
#group by is logically spliting the data
print(df.groupby('c2').sum())

#group by on multiple columns
df=pd.DataFrame([[1,2,3],[4,5,6],[7,2,3], [4,5,12]], columns=['c1','c2','c3'])
print(df)
print(df.groupby(['c2','c3']).sum())

#viewing grouped data
print(df.groupby(['c2','c3']).groups)

print(df.groupby('c2').groups)
```
