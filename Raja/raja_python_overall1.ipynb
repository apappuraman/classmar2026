**15. OOPS**

```python
#How to create class
class Student:
  def __init__(self):
    self.id=0
    self.name="unknown"

  def setId(self, id):
    self.id=id

  def setName(self, name):
    self.name=name

  def getId(self):
    return self.id

  def getName(self):
    return self.name

s=Student()
print(s.getId(), s.getName())
s.setId(101)
s.setName("Appu")
print(s.getId(), s.getName())

#inheritance
#parent child relationship
class Parent:
  def __init__(self) -> None:
    pass

  def parentFunc(self):
    print("I am in parent function without argument")

  def dispaly(self):
    print("i am in parent display...")

  def parentFunc(self, a): #python does nto support overload. It takes latest function
    print("I am in parent function with argument")


class Child(Parent):
  def __init__(self) -> None:
    super().__init__()

  def childFunc(self):
    print("I am in child")

  def dispaly(self): #override
    print("i am in child display...")

c=Child()
c.parentFunc(5)
c.childFunc()
c.dispaly()

```

**17. Interface**

```python
from abc import ABC, abstractmethod

class StudentInterface(ABC):
  @abstractmethod
  def setId(self, id):
    pass

  @abstractmethod
  def setName(self, name):
    pass

  @abstractmethod
  def display(self):
    pass

class Student(StudentInterface):
  def __init__(self):
    self.__id=0 #private valiable
    self.__name="unknown"

  def setId(self, id):
    self.__id=id

  def setName(self, name):
    self.__name=name

  def getId(self):
    return self.__id

  def getName(self):
    return self.__name

  def display(self):
    print("Student Id:", self.getId())
    print("Student Name:", self.getName())

s=Student()
#s.setId(101)
s.setName("Appu")
s.display()
```

**18. Lambda Function**

**20. Multi Threadding**

```python
from time import sleep
from threading import Thread

def mytask1():
  for i in range(10):
    print("I am in my task1")
    sleep(2)

def mytask2():
  for i in range(10):
    print("I am in my task2")
    sleep(2)

#mytask1()
#mytask2()
th1=Thread(target=mytask1)
th2=Thread(target=mytask2)
th1.start()
th2.start()

th1.join()
th2.join()

print("Main thread finished")
```

**21. Thread synchronization**

```python
from time import sleep
from threading import Thread
import threading
a=0 #shared resource between threads
lock = threading.Lock()

def myfunc1():
  global a
  global lock
  for x in range(10):
    lock.acquire()
    b=a
    sleep(1)
    b=b+1
    a=b
    lock.release()

def myfunc2():
  global a
  global lock
  for x in range(10):
    lock.acquire()
    b=a
    sleep(1)
    b=b-1
    a=b
    lock.release()
#print("Before increment and decrement", a)
#myfunc1()
#print("Intermediate increment and decrement", a)
#myfunc2()
#print("After increment and decrement", a)

print("Before thread increment and decrement", a)
th1=Thread(target=myfunc1)
th2=Thread(target=myfunc2)
th1.start()
th2.start()

th1.join()
th2.join()
print("After thread increment and decrement", a)
```
